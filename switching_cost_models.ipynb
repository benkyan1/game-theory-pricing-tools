### Analyze customer churn patterns

```python
def measure_from_churn_data(self, customer_data):
        """
        Method 1: Analyze customer churn patterns
        """
        # Calculate how price differences affect churn
        churn_rate = customer_data['churned'].mean()
        
        # Model: Churn = f(price_difference, tenure, satisfaction)
        import statsmodels.api as sm
        
        X = customer_data[['price_difference', 'tenure', 'satisfaction_score']]
        X = sm.add_constant(X)
        y = customer_data['churned']
        
        model = sm.Logit(y, X).fit()
        
        # Switching cost inferred from price difference needed to cause churn
        # If customers need $50 price advantage to switch, switching cost ≈ $50
        
        return model.params['price_difference']
```

### Direct customer surveys

```python
def measure_from_survey_data(self):
        """
        Method 2: Direct customer surveys
        """
        survey_questions = {
            'contract_penalty': "What would be the penalty for canceling your contract?",
            'setup_time': "How many hours would it take to set up with a new provider?",
            'learning_curve': "How many days to learn a new system?",
            'integration_cost': "What's the cost to migrate your data?",
            'relationship_value': "How much do you value existing relationships?"
        }
        
        # Convert survey responses to monetary equivalent
        switching_cost = (
            survey_answers['contract_penalty'] +
            survey_answers['setup_time'] * 50 +  # $50/hour
            survey_answers['learning_curve'] * 500 +  # $500/day
            survey_answers['integration_cost']
        )
        
        return switching_cost
```

### Analyze market share persistence

```python
def measure_from_market_data(self, market_data):
        # If market share is sticky despite price differences, switching costs are high
        from scipy.stats import spearmanr
        
        # Calculate correlation between price position and market share over time
        correlation = spearmanr(market_data['price_rank'], 
                                market_data['market_share_rank']).correlation
        
        # Lower correlation = higher switching costs
        switching_cost_proxy = (1 - correlation) * 100
        
        return switching_cost_proxy
```

### Analyze customer behavior patterns

```python
def measure_from_behavioral_data(self, behavioral_data):
        """
        Method 4: Analyze customer behavior patterns
        """
        # Track how many customers compare prices vs. just renewing
        price_comparison_rate = behavioral_data['compared_prices'].mean()
        auto_renewal_rate = behavioral_data['auto_renewed'].mean()
        
        # High auto-renewal = high switching costs
        switching_cost_proxy = auto_renewal_rate * 100
        
        return switching_cost_proxy
```
